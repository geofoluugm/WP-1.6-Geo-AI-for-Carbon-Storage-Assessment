```python
import os
import glob
import h5py
import pandas as pd
import geopandas as gpd
import contextily as ctx  # Untuk menambahkan basemap
```

```python
# Direktori input dan output
indir = r'E:\Folu Net Sink 2030\Lidar Gedi\Data full\Clip'
outdir = r'E:\Folu Net Sink 2030\Lidar Gedi\Data full\Shapefiles'

# Memastikan direktori output ada
os.makedirs(outdir, exist_ok=True)
```

```python
# Loop melalui setiap file H5 di direktori input
for subfile in glob.glob(os.path.join(indir, 'GEDI04_A*.h5')):
    # Reset list untuk setiap file
    lat_l = []
    lon_l = []
    agbd = []

    # Membaca file H5
    hf_in = h5py.File(subfile, 'r')
    for v in list(hf_in.keys()):
        if v.startswith('BEAM'):
            beam = hf_in[v]
            lat_l.extend(beam['lat_lowestmode'][:].tolist()) 
            lon_l.extend(beam['lon_lowestmode'][:].tolist()) 
            agbd.extend(beam['agbd'][:].tolist())
    hf_in.close()

    # Membuat GeoDataFrame untuk file ini
    geo_arr = list(zip(agbd, lat_l, lon_l))
    df = pd.DataFrame(geo_arr, columns=["agbd", "lat_lowestmode", "lon_lowestmode"])
    gdf = gpd.GeoDataFrame(df, geometry=gpd.points_from_xy(df.lon_lowestmode, df.lat_lowestmode))
    
    # Mengatur sistem koordinat (CRS) ke EPSG:4326
    gdf.set_crs(epsg=4326, inplace=True)
    
    # Mengecualikan nilai AGBD -9999
    gdf = gdf[gdf['agbd'] != -9999]

    # Memastikan ada data setelah filter
    if not gdf.empty:
        # Proyeksikan ke EPSG:3857
        gdf_epsg3857 = gdf.to_crs(epsg=3857)

        # Plotting untuk memvisualisasikan data (opsional)
        ax4 = gdf_epsg3857.plot(color='white', edgecolor='red', alpha=0.3, linewidth=5, figsize=(22, 7))
        gdf_epsg3857.plot(ax=ax4, column='agbd', alpha=0.1, linewidth=0, legend=True)

        # Menambahkan basemap dengan menggunakan EPSG langsung
        #ctx.add_basemap(ax4, crs="EPSG:3857")  # Gunakan EPSG yang jelas

    # Menentukan nama file output
    filename = os.path.basename(subfile).replace('.h5', '.shp')
    output_file = os.path.join(outdir, filename)

    # Menyimpan GeoDataFrame ke shapefile
    gdf.to_file(output_file, driver='ESRI Shapefile')

    print(f"File SHP berhasil dibuat: {output_file}")
```
